```python
from network import InterBankNetwork
from graphics import Graphics
```

# with constant initialization

```python
network = InterBankNetwork(
        n_banks=50,
        alpha_pareto=0.5,
        perc_deposit_shock=0.1,
        beta_lcr=10.0,
        beta_star_lcr=10.0,
        initial_mr=1.0,
        initial_l2s=3.0,
        collateral_value=1.0,
        init="constant",
        shock_method="dirichlet",
        constant_dirichlet=10,
    )
network.simulate(500)

Graphics(network).plot_network()
```

# with power law initialization

```python
network = InterBankNetwork(
        n_banks=50,
        alpha_pareto=0.5,
        perc_deposit_shock=0.1,
        beta_lcr=10.0,
        beta_star_lcr=10.0,
        initial_mr=1.0,
        initial_l2s=3.0,
        collateral_value=1.0,
        init="pareto",
        shock_method="dirichlet",
        constant_dirichlet=10,
    )
network.simulate(500)

Graphics(network).plot_network()
```
